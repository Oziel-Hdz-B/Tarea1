# **Calcule la norma dos**

$$\|A\|_2=\sqrt{\sum_{j=1}^n a_{i j}^2}$$

```python
import sympy as sp
def norma_2_matriz(A):
    """
    Calcula la norma 2 (norma de Frobenius) de una matriz A.

    :param A: Matriz de tamaño m x n con elementos simbólicos o numéricos.
    :return: Norma 2 de la matriz.
    """
    norma_2 = sp.sqrt(sum(A[i, j]**2 for i in range(A.shape[0]) for j in range(A.shape[1])))

    return norma_2.simplify()

if __name__ == "__main__":
        m = int(input("Introduce el número de filas de la matriz: "))
        n = int(input("Introduce el número de columnas de la matriz: "))
        print("Introduce los elementos de la matriz fila por fila, separados por espacios:")
        matriz = []
        for _ in range(m):
            fila = [sp.sympify(x) for x in input().split()]
            matriz.append(fila)

        A = sp.Matrix(matriz)
        resultado = norma_2_matriz(A)
        print(f"Norma 2 de la matriz: {resultado}")
```
